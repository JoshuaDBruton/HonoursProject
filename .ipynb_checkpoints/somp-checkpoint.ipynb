```python
import online_dictionary_learning.omp as omp
import numpy as np
import spams
from scipy import sparse
import scipy.io as spio
import matplotlib.pyplot as plt
```

```python
IP_data_mat=spio.loadmat("data/IP/Indian_pines_corrected.mat")
IP_gt_mat=spio.loadmat("data/IP/Indian_pines_gt.mat")
IP_image=IP_data_mat["indian_pines_corrected"]
IP_gt=IP_gt_mat['indian_pines_gt']

D=np.load('results/440_5_10000_dictionary.npy')
```

```python
print(D.shape)
print(IP_image.shape)
```

```python
def tile(image, ts):
    hsi = image.copy()
    tiles = np.array([hsi[i:i+ts, j:j+ts]for j in range(0, hsi.shape[1], ts) for i in range(0, hsi.shape[0], ts)])
    tiles = np.reshape(tiles, (tiles.shape[0], ts*ts, tiles.shape[3]))
    tiles = np.reshape(tiles, (tiles.shape[0]*tiles.shape[1], tiles.shape[2]))
    return np.array(tiles, dtype=np.double)
```

```python
tiles = tile(IP_image, 5)
```

```python
print(tiles.shape)
```

```python
alpha = np.zeros((tiles.shape[0], D.shape[1]))

for i, tile in enumerate(tiles):
    alpha[i] = omp.omp(D, tiles[i], L=20, eps=None)
```

```python
recon = np.dot(D, alpha.T).T
```

```python
plt.plot(tiles[21000], label='Original')
plt.plot(recon[21000], label='Reconstruction')
plt.legend()
plt.show()
```
