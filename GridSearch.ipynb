```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import scipy.io as spio
import online_dictionary_learning.omp as omp
```

```python
IP_data_mat=spio.loadmat("data/IP/Indian_pines_corrected.mat")
IP_gt_mat=spio.loadmat("data/IP/Indian_pines_gt.mat")
IP_image=IP_data_mat["indian_pines_corrected"]
IP_gt=IP_gt_mat['indian_pines_gt']

pavia_mat = spio.loadmat('data/PaviaU/PaviaU.mat')
pavia_gt_mat = spio.loadmat('data/PaviaU/PaviaU_gt.mat')
pavia_image = pavia_mat['paviaU']
pavia_gt = pavia_gt_mat['paviaU_gt']

salinas_mat = spio.loadmat('data/Salinas/salinas.mat')
salinas_gt_mat = spio.loadmat('data/Salinas/salinas_gt.mat')
salinas_image = salinas_mat['salinasA_corrected']
salinas_gt = salinas_gt_mat['salinasA_gt']
```

```python
def sparse_cluster(dictionary, image, gt, L=5,verbose=False):
    # Original shape of the image
    original_shape=gt.shape
    # flatten image and gt
    image_flat = np.reshape(image, (-1, image.shape[2]))
    gt_flat = np.reshape(gt, (-1))
    # Remove Misc. classes
    image_flat = image_flat[gt_flat!=0]
    
    # Compute Sparse Coefficients
    coefs = np.zeros((image_flat.shape[0], dictionary.shape[1]))
    
    if verbose:
        print('Computing sparse coefficients:')
    
    for i in range(coefs.shape[0]):
        coefs[i]=omp.omp(dictionary, image_flat[i], L=L, eps=None)
        if verbose:
            if i+1<image_flat.shape[0]:
                print('[' + str(i+1) + '] ' + str(np.round(((i+1)/image_flat.shape[0])*100,2)) + '%', end='\r')
            else:
                print('[' + str(i+1) + '] ' + str(np.round(((i+1)/image_flat.shape[0])*100,2)) + '%', end='\n')
                
    # KMeans
    k=np.unique(gt_flat).shape[0]-1
    kmeans = KMeans(n_clusters=k, random_state=0).fit(coefs)
    
    # Get labels
    labels = kmeans.labels_
    
    # Reconstruct image
    count, loc, n = 0, 0, gt_flat.shape[0]
    final = np.zeros(n)
    while count<labels.shape[0]:
        if (gt_flat[loc]!=0):
            final[loc]=labels[count]+1
            count+=1
        loc+=1
        
    return np.reshape(final, original_shape)
```

```python
atoms = [1000, 400, 300, 220]
sparsity = [20, 10, 5, 2]
dictionarys = []
Ls=[]

for atom in atoms:
    for sparse in sparsity:
        dictionarys.append(np.load('results/Indian_Pines/t5000_k%d_L%d.npy'%(atom, sparse)))
        Ls.append(sparse)
        
clusterings = []
for i in range(len(dictionarys)):
    clusterings.append(sparse_cluster(dictionarys[i], IP_image, IP_gt, L=Ls[i], verbose=False))
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(4,4, figsize=(15,15))
ax1[0].imshow(clusterings[0])
ax1[1].imshow(clusterings[1])
ax1[2].imshow(clusterings[2])
ax1[3].imshow(clusterings[3])
ax2[0].imshow(clusterings[4])
ax2[1].imshow(clusterings[5])
ax2[2].imshow(clusterings[6])
ax2[3].imshow(clusterings[7])
ax3[0].imshow(clusterings[8])
ax3[1].imshow(clusterings[9])
ax3[2].imshow(clusterings[10])
ax3[3].imshow(clusterings[11])
ax4[0].imshow(clusterings[12])
ax4[1].imshow(clusterings[13])
ax4[2].imshow(clusterings[14])
ax4[3].imshow(clusterings[15])
for a1, a2, a3, a4 in ax1, ax2, ax3, ax4:
    a1.axis('off')
    a2.axis('off')
    a3.axis('off')
    a4.axis('off')
    
print(np.unique(clusterings[0]))

#plt.savefig('IP_sparsity_atoms.jpg')
plt.show()
```
